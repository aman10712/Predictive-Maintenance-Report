### Here the buisness problem is that we have to predict the device malfunction date from the date its operations.
### And also to show the servicing date (25 days earlier to the failure date)

```python
Device_Data = pd.read_excel('device_data_final.xlsx')
Device_Data
```

```python
Device_Data.dtypes
```

```python
Store_data = pd.read_excel('stores_data_final.xlsx')
Store_data
```

```python
Store_data.dtypes
```

```python
Store_data['Operational_date'] = pd.to_datetime(Store_data['Operational_date'])
Store_data['Failure_date'] = pd.to_datetime(Store_data['Failure_date'])
```

```python
Store_data.dtypes
```

```python
Store_data['Working_Days'] = Store_data['Failure_date'] - Store_data['Operational_date']
```

```python
Store_data
```

```python
Store_data['Working_Days'][0].days
```

```python
def ch(x):
    return x.days
```

```python
Store_data['Working_Days'] = Store_data['Working_Days'].apply(ch)
```

```python
Store_data
```

```python
Store_data_new = Store_data[['Device_id','Working_Days']]
```

```python
final_data = Device_Data.merge(Store_data_new,on='Device_id',how='inner')
```

```python
X = final_data.drop('Working_Days',axis=1)
```

```python
X
```

```python
y = final_data[['Working_Days']]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=123)
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor,GradientBoostingRegressor,RandomForestRegressor,BaggingRegressor,ExtraTreesRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
models = [LinearRegression,Lasso,Ridge,ElasticNet,DecisionTreeRegressor,AdaBoostRegressor,GradientBoostingRegressor,RandomForestRegressor,BaggingRegressor,ExtraTreesRegressor,XGBRegressor,LGBMRegressor,SVR,KNeighborsRegressor]
```

```python
names = ['LinearRegression','Lasso','Ridge','ElasticNet','DecisionTreeRegressor','AdaBoostRegressor','GradientBoostingRegressor','RandomForestRegressor','BaggingRegressor','ExtraTreesRegressor','XGBRegressor','LGBMRegressor','SVR','KNeighborsRegressor']
```

```python
for i in range(0,len(models)):
    model = models[i]()
    model.fit(x_train,y_train)
    print(f"Mean Squared Error for Training Data of {names[i]} is {mean_squared_error(y_train,model.predict(x_train))}")
    print(f"Mean Squared Error for Testing Data of {names[i]} is {mean_squared_error(y_test,model.predict(x_test))}")
    print(f"Mean Absolute Error for Training Data of {names[i]} is {mean_absolute_error(y_train,model.predict(x_train))}")
    print(f"Mean Absolute Error for Testing Data of {names[i]} is {mean_absolute_error(y_test,model.predict(x_test))}")
```
